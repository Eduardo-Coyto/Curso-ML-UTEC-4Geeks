```python
import pandas as pd
from pandasql import sqldf
df = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")
```

```python
sqldf("select * from df limit 10")
```

```python
sqldf('SELECT location, new_cases FROM df LIMIT 10')
```

```python
sqldf("SELECT location, new_cases FROM df WHERE location = 'Uruguay'")
```

```python
sqldf("SELECT location, new_cases, date FROM df WHERE location = 'Uruguay' LIMIT 10")
```

```python
sqldf("SELECT location, new_cases, date FROM df WHERE location = 'Uruguay' ORDER BY date DESC LIMIT 10")
```
